```python
df.ocean_proximity.unique()
```

```python
def normalize_prox(x):
  if x == '<1H OCEAN':
    return '1H OCEAN'
  else:
    return x
```

```python
df.ocean_proximity = df.ocean_proximity.apply(normalize_prox)
```

```python
df.ocean_proximity.unique()
```

```python
df.head()
```

```python
categorical_columns = ['ocean_proximity']
numerical_columns =['longitude', 'latitude', 'housing_median_age', 'total_rooms', 'total_bedrooms', 'population', 'households',
                    'median_income']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder, StandardScaler
```

```python
x_train, x_test, y_train, y_test = train_test_split(df.drop('median_house_value', axis=1), df.median_house_value, test_size=0.2)
```

```python
x_train
```

```python
x_test
```

```python
preprocessor = ColumnTransformer([
    ('num', StandardScaler(), numerical_columns),
    ('cat', OneHotEncoder(), categorical_columns)
])
```

```python
from sklearn.linear_model import LinearRegression
```

```python
pipeline = Pipeline([
    ('preprocessor', preprocessor),
    ('regressor', LinearRegression())
])
```

```python
pipeline.fit(x_train, y_train)
```

```python
pipeline.score(x_train, y_train)
```

```python
pipeline.score(x_test, y_test)
```

```python
import pickle as pk
with open('model.pkl', 'wb') as f:
  pk.dump(pipeline, f)
```
